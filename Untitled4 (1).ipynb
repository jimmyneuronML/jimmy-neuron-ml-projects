```python
import pandas as pd
import matplotlib.pyplot as plt

# Titanic CSV von GitHub laden
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

# Ersten Blick in die Daten
print("Shape:", df.shape)
df.head()
```

```python
# Überlebensrate visualisieren
survival_counts = df['Survived'].value_counts()

plt.figure(figsize=(6,4))
survival_counts.plot(kind='bar', color=['red', 'green'])
plt.title("Survival Counts on Titanic")
plt.xlabel("0 = Did not survive, 1 = Survived")
plt.ylabel("Number of passengers")
plt.show()
```

```python
# Überlebensrate nach Geschlecht
gender_survival = df.groupby("Sex")["Survived"].mean()

plt.figure(figsize=(6,4))
gender_survival.plot(kind='bar', color=['blue', 'orange'])
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# Überlebensrate nach Passagierklasse
class_survival = df.groupby("Pclass")["Survived"].mean()

plt.figure(figsize=(6,4))
class_survival.plot(kind='bar', color=['purple', 'green', 'red'])
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Pclass (1 = First, 2 = Second, 3 = Third)")
plt.ylabel("Survival Rate")
plt.show()
```
